Importing Libraries

```python
from __future__ import print_function
import keras
from keras import backend as K
from keras.utils.data_utils import get_file
from keras.utils import np_utils
from keras.utils.np_utils import to_categorical
from keras.models import Sequential, Model
from keras.layers import Input, Embedding, Reshape, merge, LSTM, Bidirectional
from keras.layers import TimeDistributed, Activation, SimpleRNN, GRU
from keras.layers.core import Flatten, Dense, Dropout, Lambda
from keras.regularizers import l2, l1
from keras.layers.normalization import BatchNormalization
from keras.optimizers import SGD, RMSprop, Adam
from keras.utils import np_utils
from keras.metrics import categorical_crossentropy, categorical_accuracy
from keras.layers.convolutional import *
from keras.preprocessing import image, sequence
from keras.preprocessing.text import Tokenizer
from keras.layers.core import RepeatVector
import numpy as np
import os
from wxconv import WXC
```

```python
files = os.listdir('Stories')
num_stories =5
stories=[]
# In[read stories and join them all]
for i in range(num_stories):
    story =open('Stories/'+files[i],'rb').read()
    story = story.decode('utf-16')
    stories.append(story)
    print(files[i]+" Read!")
# In[Join]
text = ''.join(stories)
#stories = open('Opu.txt').read()
print("All Stories Read")
```

```python
b2e = WXC(order='utf2wx',lang='ben')
```

```python
e2b = WXC(order='wx2utf',lang='ben')
```

```python
e_text = b2e.convert(text)
```

```python
e_text
```

```python
b_text =e2b.convert(e_text)
```

```python
print(b_text)
```

```python
chars = e_text.split()
```

```python
chars
```

```python
vocab_size = len(chars)+1
```

```python
print('total words:', vocab_size)
```

```python
char_indices = dict((c,i) for i,c in enumerate(chars))
indices_chars = dict((i,c) for i,c in enumerate(chars))
```

```python
c_idx = [char_indices[c] for c in e_text.split()]
print("idx")
len(c_idx)
```

Here I am taking 8 words per sentance. You may change it

```python
cs =8
```

```python
c_in_dat=[]
for n in range(0,len(c_idx)-1-cs,cs):
    c_in_dat.append([c_idx[i+n] for i in range(cs)])
```

```python
c_in_dat
```

```python
x = np.stack(c_in_dat)
```

```python
x.shape
```

```python
n_fac =42
n_hidden=512


c_out_seq=[]
for i in range(1,len(c_in_dat)):
    c_out_seq.append(c_in_dat[i])
c_out_seq.append(c_in_dat[0])
```

```python
y_seq = np.stack(c_out_seq)
```

```python
y_seq.shape
```

```python
y_seq_oh = np_utils.to_categorical(y_seq)
```

```python
y_seq_oh.shape
```

Making the Sequential output Three Dimensional

```python
y_seq=np.atleast_3d(y_seq)
```

```python
y_seq.shape
```

```python
model_sequence = Sequential()
model_sequence.add(Embedding(vocab_size,n_fac,input_length=cs))
model_sequence.add(LSTM(n_hidden,return_sequences=True,activation='tanh'))
model_sequence.add(TimeDistributed(Dense(vocab_size-1,activation='softmax')))

model_sequence.compile(loss='categorical_crossentropy',optimizer=Adam())
model_sequence.summary()
```

```python
model_sequence.fit(x,y_seq_oh,batch_size=64,epochs=1000)
```
